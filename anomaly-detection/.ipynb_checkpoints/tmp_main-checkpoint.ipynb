# Split Dataset in training/validation/test

```python
# split into standard weeks
from numpy import split
from numpy import array
from pandas import read_csv
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn import preprocessing

# split a univariate dataset into train/test sets
def split_dataset(data):
    # split into standard weeks
    train, test = data[1:-328], data[-328:-6]
    #train, test = train_test_split(data, test_size=0.2, random_state=42, shuffle=False)
    # restructure into windows of weekly data
    train = array(split(train, len(train)/7))
    test = array(split(test, len(test)/7))
    return train, test

# load the new file
dataset = reader.resample(dataset)
#dataset = reader.normalize(dataset)
dataset = dataset['Global_active_power'].to_frame()

scaler = preprocessing.MinMaxScaler()
scaled_values = scaler.fit_transform(dataset.values)
dataset = pd.DataFrame(scaled_values, columns=dataset.columns, index = dataset.index)

train, test = split_dataset(dataset)
# validate train data
#print(train.shape)
# validate test
#print(test.shape)
```

```python
master_url_root = "https://raw.githubusercontent.com/numenta/NAB/master/data/"

df_small_noise_url_suffix = "artificialNoAnomaly/art_daily_small_noise.csv"
df_small_noise_url = master_url_root + df_small_noise_url_suffix
df_small_noise = pd.read_csv(
    df_small_noise_url, parse_dates=True, index_col="timestamp"
)

df_daily_jumpsup_url_suffix = "artificialWithAnomaly/art_daily_jumpsup.csv"
df_daily_jumpsup_url = master_url_root + df_daily_jumpsup_url_suffix
df_daily_jumpsup = pd.read_csv(
    df_daily_jumpsup_url, parse_dates=True, index_col="timestamp"
)
```

```python
pyplot.plot(df_daily_jumpsup)
```

```python
# Normalize and save the mean and std we get,
# for normalizing test data.
training_mean = df_small_noise.mean()
training_std = df_small_noise.std()
df_training_value = (df_small_noise - training_mean) / training_std
print("Number of training samples:", len(df_training_value))
```

```python
import numpy as np
TIME_STEPS = 288

# Generated training sequences for use in the model.
def create_sequences(values, time_steps=TIME_STEPS):
    output = []
    for i in range(len(values) - time_steps):
        output.append(values[i : (i + time_steps)])
    return np.stack(output)


x_train = create_sequences(df_training_value.values)
print("Training input shape: ", x_train.shape)
```

```python
from tensorflow import keras
from models import keras_cnn_ae
model = keras_cnn_ae.create_cnn_ae(input_shape=(x_train.shape[1], x_train.shape[2]))
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss="mse")
```

```python
#print(model.summary())
epochs = 20
batch_size = 64
validation_split = 0.1
verbose=True
history = model.fit(x_train, x_train, epochs=epochs, batch_size=batch_size, validation_split=validation_split, verbose=verbose,
    callbacks=[
        keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, mode="min")
    ],
)
```

```python
pyplot.plot(history.history["loss"], label="Training Loss")
pyplot.plot(history.history["val_loss"], label="Validation Loss")
pyplot.legend()
pyplot.show()
```

```python
# Get train MAE loss.
x_train_pred = model.predict(x_train)
train_mae_loss = np.mean(np.abs(x_train_pred - x_train), axis=1)

pyplot.hist(train_mae_loss, bins=50)
pyplot.xlabel("Train MAE loss")
pyplot.ylabel("No of samples")
pyplot.show()

# Get reconstruction loss threshold.
threshold = np.max(train_mae_loss)
print("Reconstruction error threshold: ", threshold)
```
